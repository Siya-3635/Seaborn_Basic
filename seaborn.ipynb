```python
import pandas as pd
import seaborn as sns
sales=pd.read_excel('Sample-sales-data-excel.xlsx')
sales.head()
```

```python
sns.relplot('Sales','Profit',data=sales)
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Region')
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Region',style='Region')
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Region',style='Category')
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Region',size='Discount',sizes=(50,140))
```

# line Plot

```python
sns.relplot('Order Date','Sales',data=sales,kind='line')
```

```python
sales['year']=sales['Order Date'].dt.year
sales.year.unique()
```

```python
sns.relplot('year','Sales',data=sales,kind='line')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci='sd')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',estimator=None)#estimator=sum
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style='Category',ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Category',ci=None)
```

```python
from matplotlib import pyplot as plt
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Region',ci=None)
plt.xticks(range(2014,2018))
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Region',ci=None,col_wrap=2)
plt.xticks(range(2014,2018))
```

```python
sns.catplot('Ship Mode','Sales',data=sales)
```

```python
sns.catplot('Ship Mode','Sales',data=sales,jitter=False)#this chart is shows the magnitude
```

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit>800'),jitter=False)
```

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit>100'),kind='swarm')
```

```python
sns.catplot('Sales','Ship Mode',data=sales.query('Profit>100'),kind='swarm')
```
